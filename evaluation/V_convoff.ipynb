```python
def V(x):
    return x / 1023. * 1800

def I(x):
    return x / 1023. * 2500

def cumcharge(current, time):
    return integrate.cumtrapz(current, time, initial=0)

def charge(current, time):
    return numpy.diff(cumcharge(current, time))

colors = ['red', 'blue', 'green', 'black', 'cyan', 'orange', 'pink', 'yellow', 'grey']
```

```python
def plot_effect_of_v_convoff(nrn = default_nrn, experiment=experiment):
    V_convoffi = V(experiment.get_parameters_and_results(nrn, (np.V_convoffx, ), tuple())[:,0])
    v_convoffi = [M.traces[nrn][NETS_AND_PINS.V_synx_integrator_reference][-1] for M in experiment.measurements]
    v_synint = [M.traces[nrn][NETS_AND_PINS.V_synx_integrator][-1] for M in experiment.measurements]

    figure = bk.figure(
        plot_width=800,
        plot_height=600,
        title="Effect of V_convoff",
        tools=TOOLS,
        x_axis_label="flaoting gate: V_convoffi[V}",
        y_axis_label="voltage on OTA[V]",
    )

    figure.scatter(V_convoffi, v_convoffi, color="navy")
    figure.scatter(V_convoffi, v_synint, color="red")
    bk.show(figure)
    
def plot_membranes_with_spikes(nrn = default_nrn, experiment=experiment):
    # Create a set of tools to use
    
    DATA = [
        numpy.array([M.traces[nrn]['t'] for M in experiment.measurements]),
        numpy.array([M.traces[nrn]['V'] for M in experiment.measurements]),
        numpy.array([numpy.mean(M.traces[nrn][NETS_AND_PINS.V_synx_integrator_reference]) for M in experiment.measurements]),    
    ]
    DATA.extend(experiment.get_parameters_and_results(nrn, (np.V_convoffx, ), ('baseline',)).T)

    figure = bk.figure(
        plot_width=800,
        plot_height=600,
        title="Neuron traces with spike input",
        tools=TOOLS,
        x_axis_label="simulation time [s]",
        y_axis_label="membrane voltage [mV]",
    )

    for ii, data in enumerate(zip(*DATA)):
        t, v, ota_n, v_convoff, baseline = data
       
        text="{}/{:.0f}/{:.0f}".format(ii, ota_n * 1.0e3, V(v_convoff))
        
        dt = numpy.linspace(t[-1], 0, 25)
        line = figure.line(t, v * 1.0e3, color='navy')
        figure.text(dt[ii%8], v[-1] * 1.0e3, text=text,
                    text_baseline='bottom', text_align="right", 
                    text_color="darkred", text_font_size="8pt")
        figure.line([t[0], t[-1]], [baseline * 1.0e3, baseline * 1.0e3], color='red', line_dash="8 8")

        
    data = V(experiment.get_parameters_and_results(nrn, (np.E_l, np.E_synx), tuple()))
    E_l, E_synx  = numpy.mean(data, axis=0)
    t = t[0], t[-1]
    figure.line(
        t, [E_l, E_l], color='black', line_dash="8 8",
        legend="resting potetial")
    # figure.line(
    #     t, [E_synx, E_synx], color='black', line_dash="4 4",
    #     legend="revesal potetial")
    
    bk.show(figure)

def plot_integral_over_baseline(nrn = default_nrn, experiment=experiment):

    DATA = experiment.get_parameters_and_results(nrn, (np.V_convoffx, np.E_synx), ('psp_area', 'baseline'))
    V_convoffx, E_synx, psp_area, baseline = DATA.T
    
    psp_area = psp_area / (V(E_synx)/1e3 - baseline)
    
    figure = bk.figure(
        plot_width=800,
        plot_height=600,
        title="area of PSP",
        tools=TOOLS,
        x_axis_label="V_convoffx [mV]",
        y_axis_label="area of psp / distance to reversal potential [mVs]",
    )
    figure.scatter(psp_area, baseline, color='navy', legend="PSP areas")
    for label, x, y in zip(V_convoffx, psp_area, baseline):
        figure.text(x, y, text="{}".format(label),
                    text_baseline='bottom', text_align="left", 
                    text_color="darkred", text_font_size="8pt")
    bk.show(figure)


def plot_psp_height_vs_leak_current(nrn = default_nrn, experiment=experiment):

    DATA = experiment.get_parameters_and_results(nrn, (np.V_convoffx, ), ('psp_area', 'syn_resting_current'))

    figure = bk.figure(
        plot_width=800,
        plot_height=600,
        title="area of PSP",
        tools=TOOLS,
        x_axis_label="leakage of OTA [nA]",
        y_axis_label="area of psp [mVs]",
    )
    figure.scatter(DATA[:,2] * 1e9, DATA[:,1] * 1e3, color='navy', legend="PSP areas")
    for label, y, x in DATA:
        figure.text(x * 1e9, y * 1e3, text="{}".format(V(label)),
                    text_baseline='bottom', text_align="left", 
                    text_color="darkred", text_font_size="8pt")
    bk.show(figure)
    
def plot_output_current_of_synaptic_input(nrn = default_nrn, experiment=experiment):
       
    v_confoff, I_ota, I_syn = experiment.get_parameters_and_results(
        nrn, (np.V_convoffx, ), ("ota_resting_current", "syn_resting_current")).T
    
    figure = bk.figure(
        plot_width=800,
        plot_height=600,
        title="Synaptic Current",
        tools=TOOLS,
        x_axis_label="V_convoff [mV]",
        y_axis_label="Current [nA]",
    )


    figure.scatter(V(v_confoff), I_ota * 1e9, color='maroon',
                       legend="OTA output")
    figure.scatter(V(v_confoff), I_syn * 1e9, color='navy',
                       legend="Synaptic output")
    bk.show(figure)
    
def plot_result_fitting(nrn = default_nrn, experiment=experiment):

    DATA = experiment.get_parameters_and_results(
        nrn,
        (np.V_convoffx, np.E_synx),
        ('psp_area', 'baseline', "syn_resting_current"))
    # x, y = V(DATA[0]),  DATA[2] / (V(DATA[1]) - DATA[3])
    V_convoff, E_synx, area, baseline, syn_leak = DATA.T
    area = area / (V(E_synx)/1e3 - baseline)
    V_convoff = V(V_convoff)
    
    figure = bk.figure(
        plot_width=800,
        plot_height=600,
        title="area of PSP",
        tools=TOOLS,
        x_axis_label="V_convoffx [DAC]",
        y_axis_label="normalized value",
    )
    
    # popt, pcov = curve_fit(piece_wise_linear, x, y, p0=initial_guess(x, y))
    
    def normalize(x):
        a, b = numpy.min(x), numpy.max(x)
        return (x - a) / (b - a)
    
    area_n = normalize(area) - 1.0
    baseline_n = normalize(baseline)
    figure.scatter(V_convoff, area_n, color='navy', legend="normalized PSP areas")
    figure.scatter(V_convoff, baseline_n, color='red', legend="normalized baseline")
    figure.scatter(V_convoff, area_n + baseline_n, color='mediumseagreen', legend="sum of both")
    
    d = baseline_n - area_n
    w = numpy.max([(-area_n)/(baseline_n - area_n), (baseline_n)/(baseline_n - area_n)], axis=0)
    
    f = numpy.poly1d(numpy.polyfit(V_convoff, area_n + baseline_n, 5, w=w))
    figure.line(V_convoff, f(V_convoff), color='black', alpha=0.7)
    # figure.scatter(V_convoff, w, marker='diamond', color='black', alpha=0.6)
    bk.show(figure)
    
    r = f.roots[numpy.isreal(f.roots)]
    r = r[(V_convoff[0] < r) & (r < V_convoff[-1])]
    r = r.real
    
    figure = bk.figure(
        plot_width=800,
        plot_height=600,
        title="Synaptic leakage current",
        tools=TOOLS,
        x_axis_label="V_convoffx [DAC]",
        y_axis_label="current [nA]",
    )
    figure.scatter(V_convoff, syn_leak)
    figure.line([r[0], r[0]], [numpy.min(syn_leak), numpy.max(syn_leak)])
    bk.show(figure)
    
def plot_results(results=results):
    
    from pycake.calibrator import V_convoffx_Calibrator
    
    calib = V_convoffx_Calibrator(experiment)
    x = dict(calib.generate_coeffs())
    data = x[np.V_convoffx]
    
    from bokeh.charts import Histogram
    from collections import OrderedDict
    
    hist_data = OrderedDict(V_convoffx=V(numpy.array(data.values())))
    hist = Histogram(
        hist_data, bins=50,
        title="Distribution of V_convoffx",
        ylabel="frequency",
        xlabel="V_convoffx [mV]",
        legend="top_left",
        width=800, height=800, notebook=True)
    bk.show(hist)
    
    ota_leak = []
    syn_leak = []
    for neuron, result in data.iteritems():
        x, y, z = experiment.get_parameters_and_results(nrn, (np.V_convoffx, ), ("ota_resting_current", "syn_resting_current", )).T
        ota_leak.extend(numpy.interp(result, x, y))
        syn_leak.extend(numpy.interp(result, x, z))
    ota_leak = numpy.array(ota_leak) * 1e9
    syn_leak = numpy.array(syn_leak) * 1e9
    
    hist_data = OrderedDict()
    hist_data["OTA leakage"] = ota_leak
    hist_data["synaptic leakage"] = syn_leak
    hist = Histogram(
        hist_data, bins=50,
        title="Distribution of leakage currents",
        ylabel="frequency",
        xlabel="leakage [nA]",
        legend="top_left",
        width=800, height=800, notebook=True)
    bk.show(hist)
```

# Goals of V_convoff(i/x) calibration

* compensating the fixed-pattern(?) offset of the OTA in the synaptic input
* finding the optimal setting:
    * Setting V_confoff to large, effectivly supresses spikes
    * Setting V_confoff to small, causes permanent leakage to the membrane

### Spike based method (it works)
Idea:
* maximise the size off an PSPs

Problems:
* PSPs size also depends on other parameters (V_syntx, I_gl, biases, ...)

Results:
* for (bias=0nA) for V_confoffi < 970mV the synaptic input pulls the membrane to the reversal potential
* for (bias>0nA) the translation is strongly smeared out, no reliable evaluation possible
* Seams feasible, when only the V_convoffi floating gate is updated

```python
def print_srams(experiment=experiment):
    t_field = "{{:>{w}}}"
    
    SRAMS = (
        [getattr(NETS_AND_PINS, "V_synapse_sram_top{}".format(ii)) for ii in range(4)],
        [getattr(NETS_AND_PINS, "V_synapse_sram_bottom{}".format(ii)) for ii in range(4)],
    )

    def get_fields(M):
        return ["".join(str(int(numpy.mean(M.traces[nrn][net]) > 0.9)) for net in RAM) for RAM in SRAMS]
        
    
    fields = ('top', 6), ('bottom', 6)
    header = "|     | " + " | ".join(t_field.format(w=f[1]).format(f[0]) for f in fields) + " |"
    print header
    print len(header) * '-'
    
    for ii, M in enumerate(experiment.measurements):
        f = [t_field.format(w=3).format(ii)]
        f.extend(t_field.format(w=w[1]).format(v) for w, v in zip(fields, get_fields(M)))
        print "| " + " | ".join(f) + " |"

def plot_synaptic_currents(run, experiment=experiment):
    nrn = NeuronOnHICANN(Enum(1))
    t_range = (2.995e-5, 3.010e-5)    
    
    NORMAL = "3 3"
    DASHED = "4 8"
    plt = [
        ('red', NORMAL, NETS_AND_PINS.I_synapse_out_exc),
        ('red', DASHED, NETS_AND_PINS.I_synapse_out_inh, ),
        ('navy', NORMAL, NETS_AND_PINS.I_synapse_in_top, ),
        ('navy', DASHED, NETS_AND_PINS.I_synapse_in_bottom, ),
        ('seagreen', NORMAL, NETS_AND_PINS.I_synapse_vgmax_top, ),
        ('seagreen', DASHED, NETS_AND_PINS.I_synapse_vgmax_bottom, ),
        ('indigo', NORMAL, NETS_AND_PINS.I_synapse_Iout_top, ),
        ('indigo', DASHED, NETS_AND_PINS.I_synapse_Iout_bottom, ),
    ]
    
    M = experiment.measurements[run]
    
    figure = bk.figure(
        plot_width=800,
        plot_height=600,
        title="Synaptic Currents",
        tools=TOOLS,
        x_axis_label="time [s]",
        y_axis_label="current [nA]",
    )
    
    t = M.traces[nrn]['t']
    idx = slice(t.searchsorted(t_range[0]), t.searchsorted(t_range[1]))
    for color, line_dash, net in plt:
        Y = M.traces[nrn][net] * 1e9
        line = figure.line(t[idx], Y[idx], color=color, line_dash=line_dash, legend=net, alpha=0.6)
    bk.show(figure)
    
def plot_synaptic_voltages(run, experiment=experiment):
    nrn = NeuronOnHICANN(Enum(1))
    t_range = (2.995e-5, 3.010e-5)    
    
    NORMAL = "5 3"
    DASHED = "8 2"
    plt = [
        ('red', NORMAL, NETS_AND_PINS.V_synapse_pre_top),
        ('navy', NORMAL, NETS_AND_PINS.V_synapse_eni_top, ),
        ('seagreen', NORMAL, NETS_AND_PINS.V_synapse_enx_top, ),
        ('indigo', NORMAL, NETS_AND_PINS.V_synapse_enout_top, )
    ]
    
    M = experiment.measurements[run]
    
    figure = bk.figure(
        plot_width=800,
        plot_height=600,
        title="Synaptic Voltages top",
        tools=TOOLS,
        x_axis_label="time [s]",
        y_axis_label="voltage [mV]",
    )
    
    t = M.traces[nrn]['t']
    idx = slice(t.searchsorted(t_range[0]), t.searchsorted(t_range[1]))
    for color, line_dash, net in plt:
        Y = M.traces[nrn][net] * 1e3
        line = figure.line(t[idx], Y[idx], color=color, line_dash=line_dash, legend=net, alpha=0.6, line_width=3)
    bk.show(figure)
    
    plt = [
        ('red', DASHED, NETS_AND_PINS.V_synapse_pre_bottom, ),
        ('navy', DASHED, NETS_AND_PINS.V_synapse_eni_bottom, ),
        ('seagreen', DASHED, NETS_AND_PINS.V_synapse_enx_bottom, ),
        ('indigo', DASHED, NETS_AND_PINS.V_synapse_enout_bottom, ),
    ]
    
    M = experiment.measurements[run]
    
    figure = bk.figure(
        plot_width=800,
        plot_height=600,
        title="Synaptic Voltages bottom",
        tools=TOOLS,
        x_axis_label="time [s]",
        y_axis_label="voltage [mV]",
    )
    
    t = M.traces[nrn]['t']
    idx = slice(t.searchsorted(t_range[0]), t.searchsorted(t_range[1]))
    for color, line_dash, net in plt:
        Y = M.traces[nrn][net] * 1e3
        line = figure.line(t[idx], Y[idx], color=color, line_dash=line_dash, legend=net, alpha=0.6, line_width=3)
    bk.show(figure)
    
#plot_synaptic_currents(1)
#plot_synaptic_voltages(1)
# ot_synaptic_currents(2)
```
